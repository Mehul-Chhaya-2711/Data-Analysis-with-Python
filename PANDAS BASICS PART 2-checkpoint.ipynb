```python
#ALL ABOUT DATA WRANGLING, CLEANING, TRANSFORMING, RESHAPING AND MERGING

#below imports are done to facilitate better outputs as well as pandas and numpy operations through out the notebook
from IPython.display import display, HTML ,display_html
from pandas import Series, DataFrame
import pandas as pd
import numpy as np

CSS = """
.output {
    flex-direction: row;
}
"""

HTML('<style>{}</style>'.format(CSS))

def display_side_by_side(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

```python
#MERGING DATA SETS

#Example Dataframe

df1 = DataFrame({'Name':['Mehul','Monish','Jayash','Archi','Megha','Krunal','Rajat','Pashmina','Naman','Merath','Rachit'],
                'City':['Ahmedabad','Ahmedabad','Sydney','Ahmedabad','Mumbai','Valsad','Dhhrangadhhra','Dubai','Ahmedabad','Bangalore','Gandhinagar'],
               'Phone':[123,23,345,567,867,345,234,321,234,243,666]})
df2 = DataFrame({'Name':['Mehul','Monish','Jayash','Archi','Megha','Krunal','Rajat','Pashmina','Naman','Merath','Aritra'],
              'Salary':[1,2,3,4,5,6,7,7,None,None,None],
             'Married':[1,0,1,0,0,0,1,0,0,1,1]})
df3 = DataFrame({'Surname':['Chhaya','Shah','Mistry','Patel','Bharti','Sonavane','Oza','Babaycone','Sherasiya','Chhaya','Chhaya'],
                   'Phone':[123,23,345,567,867,345,234,321,234,243,666]})
df4 = DataFrame({'Christianed_as':['Mehul','Monish','Jayash','Archi','Megha','Krunal','Rajat','Pashmina','Naman','Merath','Aritra'],
              'Salary':[1,2,3,4,5,6,7,7,None,None,None],
             'Married':[1,0,1,0,0,0,1,0,0,1,1]})
print('Normal Example Dataframe')
display_side_by_side(df1,df2,df3)
```

```python
print('Merging Dataframe Without Specifying keys : Merge will happen on the common column')
print('When Not specifying keys it is compulsory to have a common columns between 2 data frames to merge and avoid error')
display(pd.merge(df1,df2))
```

```python
print('Merging by specifying the key/column names for left as well as right table')
print('Merge by default takes inner join unless specified')
display(pd.merge(df1,df4,left_on='Name',right_on='Christianed_as'))
```

```python
print('Merge by applying specific joining logic such as left, right, inner or outer')
display_side_by_side(pd.merge(df1,df4,left_on='Name',right_on='Christianed_as',how='left'),pd.merge(df1,df4,left_on='Name',right_on='Christianed_as',how='right'),pd.merge(df1,df4,left_on='Name',right_on='Christianed_as',how='outer'))
```

```python
print('Merging on multiple keys || Just like joining in SQL with more than 1 join conditions')
display(pd.merge(df1,df4,left_on = ['Name','Phone'],right_on = ['Christianed_as','Married'],how = 'outer'))
```

```python
#merging on index
print('Merging on indexes gives us a Rachit and Aritra on same row where as same was not the case perviously')
print('This happens because the join is not happening on Names but index positions')
display_side_by_side(pd.merge(df1,df4,left_index = True,right_index = True,how = 'outer'),pd.merge(df1,df4,left_on = 'Name',right_on = 'Christianed_as',how = 'outer'))
print('Getting Surnames')
display(pd.merge(df1,df3,left_index = True,right_index = True))
```

```python
#USING CONCATENATION
#REMEMBER; THE CONCATENATION WORKS ALONG COLUMNS I.E SERIES. IT IS A NUMPY FUNCTIONALITY. 
#WE CAN USE CONCATENATION OPERATION OVER COLUMNS TO MATERIALIZE JOINS/MERGE OPERATIONS
```

```python
# FOR THE DATA FRAMES IN ABOVE EXAMPLE; LETS JOIN BY NAMES
# IF WE NEED TO PRODUCE A UNION ALL I.E. DATAFRAME OVER DATAFRAME KIND OF RESULT; AXIS = 0
# IF WE NEED TO PRODUCE A CROSS TABULATED RESULT I.E A SINGLE DATAFRMAE; AXIS = 1
df5 = DataFrame({'Name':['Mehul','Monish','Jayash','Archi','Megha','Krunal','Rajat','Pashmina','Naman','Merath','Aritra'],
               'Salary':[1,2,3,4,5,6,7,7,None,None,None],
              'Married':[1,0,1,0,0,0,1,0,0,1,1]})
display_side_by_side(pd.concat([df1,df4]),pd.concat([df1,df4],axis = 1 ,keys = [1,2]))  #keys added here are nothing by hierarchical indexing
print('Difference between the 2 outputs is that values of same column names gets arranged automatically in same line')
display_side_by_side(pd.concat([df1,df5],ignore_index=True),pd.concat([df1,df5]))  #Notice the differenc ein indexing
# Why do we need hierarchical indexing? It helps us differentiate between part obtained from a table and another in the final result
```

```python
#Let us not allow duplicates
#As we can see first column i.e. index itself has duplicate values will throw Exception upon concat operation
#To overcome the same we use ignore_index = True
#To not allow duplicates we can keep using verify_integrity=True
try:
    display(pd.concat([df1,df5],verify_integrity=True))
except Exception as e:
    print(e)
    display_side_by_side(pd.concat([df1,df5],ignore_index = True,verify_integrity=True),pd.concat([df1,df5],axis = 1))
    
#Thing to remember : Concat works on axis' indexes; one can provide desired values of indexes but remember it does not 
#work on columns/keys as does merge/join
```

```python
#RESHAPING AND PIVOTING DATAFRAMES
#THIS IS WHERE HIERARCHICAL INDEXING COME IN PICTURE
```

```python
#to give row names/labels use index key word; for columns it will be columns only

data = DataFrame(np.arange(6).reshape((2, 3)),index=pd.Index(['Ohio', 'Colorado'], name='state'),
                 columns=pd.Index(['one', 'two', 'three'], name='number'))
#State is the name of row indexes and Number is the name of column indexes
display('Original Data Frame',data)
columns_2_rows = data.stack() 
display('Columns brought to rows',columns_2_rows) #Results produced here are not data frames but series with multiple hierarchical indexes
display('Re-generated the original Data Frame',columns_2_rows.unstack()) #produces the original data frame
print('Getting only the state/columns values to columns')
print('When unstacking in a DataFrame, the level unstacked becomes the lowest level in theresult')
display_side_by_side(columns_2_rows.unstack('state'),columns_2_rows.unstack('number'))


#what happened here?
# 1. We got columns to rows by stacking. Meaning, State index remained as is with an additional index of one,two&three added
# 2. Now we have a series of hierarchical data frame
# 3. Now we perform unstacking. Absolute instacking provides us back the original data frame
# 4. Unstacking can also be done only according to any of the multiple hierarchical indexes
# such as IF we do it by states; states will go to column, if we do it by number one,two&three goes to columns
```

```python
# Stacking filters out missing data by default, so the operation is easily invertible:
s1 = Series([0, 1, 2, 3], index=['a', 'b', 'c', 'd'])
s2 = Series([4, 5, 6], index=['c', 'd', 'e'])
data2 = pd.concat([s1, s2], keys=['one', 'two'])
display(data2)
display(data2.unstack())
display('Stacking filters out missing data by default',data2.unstack().stack())
display(data2.unstack().stack(dropna=False))
```

```python
#REGULAR EXPRESSIONS
import re
text = """Dave dave@google.com
Steve steve@gmail.com
Rob rob@gmail.com
Ryan ryan@yahoo.com
"""

# We want to search email ids from this
# So pattern will be  Aplha_Numeric_symbols + @ + Aplha_Numeric_symbols +. + Aplha_Numeric

pattern = r'[a-z0-9._%+-]+@[a-z0-9.-]+\.[a-z]{2,4}'

regex = re.compile(pattern,flags = re.IGNORECASE)

#Find all the occurrances of pattern
print(regex.findall(text))
#Find only the first occurance of pattern
print(regex.search(text))
#Checks if the pattern occurs at the start of the string else returns None
print(regex.match(text))
#use the pattern to get parts of an email_id

pattern = r'([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})'
regex = re.compile(pattern,flags = re.IGNORECASE)

parts = regex.match('abc@def.com')
print(parts.groups())
print(regex.findall(text))
```

```python
#A SIMPLE DATA CRUNCHING WITH FOOD DATABASE: THERE IS NO SPECIFIC PURPOSE THAT IS BEING SERVED HERE. A DETAILED APPROACH
# TO SUCH DATA CRUNCHING CAN BE LOOKED AT AFTER REFERRING TO PANDAS BASICS PART 3
import json
database = json.load(open('D:\Study Materials\Documents to play with in python\\foods_database.json'))  #a list it is
#The output list from json file is a collection of dictionaries with keys as columns and values as row entries
```

```python
keys = database[0].keys()  #this gives us the keys (columns names of our 'to be' data frame)
print(keys)
database[0]['nutrients'][0]
#Each key further has 4 values in it under a dictionary
```

```python
#lets make a nutrients dataframe
Nutrients = DataFrame(database[0]['nutrients'])
Nutrients[:10]
```

```python
#let us make a information database
Info = DataFrame(database,columns = ['description','group','id','manufacturer'])
Info[:10]
```

```python
#Let us see the distribution of records across groups in info table
pd.value_counts(Info.group)   #value count' argument has to be a series in our case which will be a column of dataframe Info
```

```python
#Now let us form a single large table out of all the information we have

# Steps:
# 1. Convert each list(Which is a dictionary of key value pairs) of food nutrients in to a data frame
# 2. Add a column for the food id
# 3. Append the dataframe to a list
# 4. Perform above step by concat

Nutrientss = []
for records in database:
    nutrients = DataFrame(records['nutrients'])
    nutrients['id'] = records['id']
    Nutrientss.append(nutrients)

Nutrientss = pd.concat(Nutrientss, ignore_index=True)
```

```python
# Lets check if there are duplicates

print(Nutrientss.duplicated().sum()) #Nutrients.duplicated() will give us a True False sequence by records being duplicate. Sum will get us number of such duplicates

#lets remove the duplicates
Nutrientss.drop_duplicates()
```

```python
#Let us rename description to Foods and Group to Fgroup in Nutrients dataframe that has only Nutrients values

Info = Info.rename(columns = {'description':'food','group':'fgroup'})
Nutrientss = Nutrientss.rename(columns = {'description':'nutrient','group':'nutgroup'})
```

```python
#Let us perform the merge operation now
ndata = pd.merge(Nutrientss, Info, on='id', how='outer')
```

```python
# Plot median values by food group

result = ndata.groupby(['nutrient','fgroup'])['value'].quantile(0.5)
result['Zinc, Zn'].sort_values().plot(kind='barh')
```
